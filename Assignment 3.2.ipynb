### Load the dataframe from file

```python
neigh_df = pd.read_csv('df.csv')
neigh_df.head()
```

```python
neigh_df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
neigh_df
```

### Loading latitude and longitude values from csv

```python
latlng_df = pd.read_csv('Geospatial_Coordinates.csv')
latlng_df.rename(columns = {'Postal Code': 'PostalCode'}, inplace = True)
latlng_df.head()
```

### Merge the current DF with geo-coordinates DF

```python
neigh_merge_df = pd.merge(neigh_df, latlng_df, on='PostalCode', how='left')
neigh_merge_df.head()
```

#### export to file the dataframe

```python
neigh_merge_df.to_csv('neigh_merge.csv')
```
